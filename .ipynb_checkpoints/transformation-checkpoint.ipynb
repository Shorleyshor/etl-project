```python
# Importing Dependencies
import pandas as pd
from sqlalchemy import create_engine
```

# Store CSV into DataFrame

```python
# Creating csv path and loading first data in Pandas DataFrame
tb_rates_csv = "/Users/ola/Desktop/etl-project/Resources/tb_rates.csv"
tb_rates_df = pd.read_csv(tb_rates_csv)
tb_rates_df.head()
```

```python
# Creating csv path and loading second data set in Pandas DataFrame
csv_path = "/Users/ola/Desktop/etl-project/Resources/BCG_vaccine.csv"
vaccine_df = pd.read_csv(csv_path)
vaccine_df.head()
```

#  Clean DataFrame

```python
# Dropping/removing unwanted columns
tb_rates_df.drop(["National data", "Report type", "ASR", "Confirmed cases"], axis=1, inplace=True)
```

```python
# Dropping/removing unwanted rows
new_tb_rates_df = tb_rates_df.drop([27, 29], axis=0)
new_tb_rates_df.head()
```

```python
# Renaming columns headers

mapping = {new_tb_rates_df.columns[0]:'country', new_tb_rates_df.columns[1]: '2012_cases', new_tb_rates_df.columns[2]:'2012_rate', new_tb_rates_df.columns[3]: '2011_cases', new_tb_rates_df.columns[4]:'2011_rate', new_tb_rates_df.columns[5]: '2010_cases',new_tb_rates_df.columns[6]:'2010_rate', new_tb_rates_df.columns[7]: '2009_cases',new_tb_rates_df.columns[8]:'2009_rate', new_tb_rates_df.columns[9]: '2008_cases',new_tb_rates_df.columns[10]: '2008_rate'}
transforme_data = new_tb_rates_df.rename(columns=mapping)

transforme_data.head()
```

```python
# Creating a filtered dataframe from specific columns

vaccine_cols = ['Country / Region', '2012', '2011', '2010', '2009', '2008']
vaccine_transformed= vaccine_df[vaccine_cols].copy()


# Rename the column headers
vaccine_transformed = vaccine_transformed.rename(columns={"Country / Region": "country"})

# Clean the data by dropping duplicates and setting the index
vaccine_transformed.drop_duplicates()
#vaccine_transformed.set_index("id", inplace=True)

vaccine_transformed.head()
```

```python
# Dropping NaN 

vaccine_transformed.dropna()
```
